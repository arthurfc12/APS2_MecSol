```python


T = np.zeros((6,6))
T_ = np.zeros((6,6))

for i in range(6):
    T[0][i]=100
    T_[0][i]=100


print("\nTemperaturas iniciais na barra: \n")
matprint(T)
    

lines = T.shape[0]
columns = T.shape[1]

alpha = 0.25 #1e-4

delta_t = 1e-2
delta_x = 0.1
tempo = 0
tempo_max = 10

F0 = alpha*delta_t/(delta_x**2)




while tempo <= tempo_max:
    for i in range(1,lines-1):
        for j in range(1,columns-1):
            T_[i][j] = F0*(T[i+1][j] + T[i-1][j] + T[i][j+1] + T[i][j-1]) + (1 - 4*F0)*T[i][j]
            
    T = np.copy(T_)
    tempo += delta_t


print("\nTemperaturas finais na barra: \n")
matprint(T)
```
